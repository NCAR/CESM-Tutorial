## CAM-chem and WACCM component sets

The CESM3 components can be combined in numerous ways to carry out various scientific or software experiments. A particular mix of components, along with component-specific configuration and/or namelist settings, is called a component set or ``compset``. 

**WACCM** (Whole Atmosphere Community Climate Model) and **CAM-chem** (Community Atmosphere Model with chemistry) are atmospheric components of CESM, like CAM. They can be run as ``F`` and ``B`` compsets, as well as with nudged meteorology (for example towards reanalysis). They can be run using various dycores and resolutions.
In contrast to CAM, WACCM and CAM-chem run with different possibilities of chemistry complexity, while CAM uses simplified chemistry. CAM and CAM-chem use the same vertical resolution and model top.
Different chemical mechanisms for CAM-chem with more or less complexity have been developed based on the MOZART-T1S mechanism. More comprehensive information about chemistry mechnisms is on https://www2.acom.ucar.edu/gcm/mozart.
WACCM has a model top at about 150km and therefore covers the Troposphere, Stratosphere, Mesosphere and Lower-Thermosphere. The most comprehensive WACCM mechanism is therefore called the T1MA (previously called TSMLT) chemical mechanism.
WACCM can also run with Middle Atmosphere (MA) chemistry (e.g., FHISTC_WAma, which includes very simplified chemistry in the troposphere. WACCM can also be run with Specified Chemistry (e.g., FHISTC_WAsc), similar to the CAM specified oxidant chemistry.


## CAM-chem and WACCM F component sets

In CESM3 we are in the process of scientifically validating model setup, in preparation for CMIP7 simulations.
The model may be run in free-running mode, or nudged to meteorological fields from reanalysis (U, V, T), for example from MERRA2 or ERA5. Reanalysis is regridded to the model horizontal and vertical resolution. Nudging within CAM7 currently requires changes to namelist variables.

## Overview of the Challenge Exercises for CAM-chem/WACCM
Start running one control case and choose between two options:

a) Run a CAM-chem (Low-Top) compset with T1S chemistry, historical SSTs and ~2 degree horizontal resolution <br>
b) Run a WACCM with T1MA chemistry, historical SSTs and ~2 degree horizontal resolution

Once a control case has been configured and run, perform two test cases - one changing chemistry and one changing emissions.
