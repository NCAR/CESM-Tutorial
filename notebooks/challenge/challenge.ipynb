# Challenge Exercises

The purpose of this section is to help you apply the concepts and skills introduced throughout the CESM tutorial.

The challenge exercises are intentionally less prescriptive than the previous lessons. Rather than following a detailed set of instructions, you will use your understanding of CESM to design, configure, run, and analyze your own experiments.

Each challenge exercise focuses on a different aspect of the model and may involve:

- Creating and modifying CESM cases
- Exploring different compsets and resolutions
- Modifying namelist settings
- Changing input datasets
- Making source-code modifications
- Comparing simulations and interpreting model output

Feel free to complete all of the challenge exercises or focus on the ones most relevant to the CESM components and scientific questions that interest you.

The goal is not only to obtain a successful model run, but also to develop confidence in navigating CESM, understanding its configuration options, and designing your own modeling experiments.
