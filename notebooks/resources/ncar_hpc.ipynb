# NCAR Supercomputer

Once you have an account and the necessary software, you can log in and run jobs on the NCAR supercomputer. The NCAR supercomputer is also referenced to the NCAR High Performance Computing (HPC). 

___
## Logging in on an NCAR system

To log in, start your terminal or Secure Shell client and run an `ssh` command as shown here:


```
ssh -XY username@derecho.hpc.ucar.edu 
```
OR 
```
ssh -XY username@casper.hpc.ucar.edu
```


When you log in to the NCAR HPC machines, we recommend you use `X11` forwarding. `X11` forwarding is an SSH protocol that enables users to run graphical applications on a remote server and interact with them using their local display and I/O devices. It is frequently used by CESM developers for securely interacting with NCAR HPC remote machines.

You can do this by adding the `-XY` option when logging into the machines with `SSH`.

## Using Casper

The Casper cluster is a system of specialized data analysis and visualization resources. Casper is not used for building or running CESM, but it is used for CESM data analysis. We will use Casper for the diagnostics section of the lab activities. To utilize the **_additional_** analysis tools described in the [analysis tools section](https://ncar.github.io/CESM-Tutorial/notebooks/diagnostics/additional/analysis_tools.html) you will also want to use Casper.

## Using JupyterHub

The diagnostics section of this tutorial will use JupyterHub to run Jupyter Notebooks. You will need to login to the NCAR JupyterHub system to run the notebooks successfully. We will use Open OnDemand to launch JupyterHub.

**Step 1. Access and bookmark the NSF NCAR Open OnDemand portal at: [ondemand.hpc.ucar.edu](https://ondemand.hpc.ucar.edu)**

**Step 2. Upon visiting the NSF NCAR Open OnDemand web portal, you will need to login and authenticate with your CIT credentials**

**Step 3. Navigate to Jupyter under Interactive Apps**

![Open OnDemand Landing Page](../../images/resources/OnDemand_interactive_apps.png)
*<p style="text-align: center;"> Figure: Open OnDemand Landing Page. </p>*

**Step 4. Select the appropriate resources for your job**

![Entering options for server on JupyterHub](../../images/resources/OnDemand_launch_Jupyter_batch.png)
*<p style="text-align: center;"> Figure: Entering options for server on JupyterHub. </p>*
Select the “Casper PBS Batch” option in the drop down menu. After, you will need to make other choices to select the right resources: 3 hours wall time to ensure your server is active for the entire hands-on activity session and Project Account `UESM0016`. Now launch the server.

**Step 5. Connect to Jupyter**

![Launching Jupyter](../../images/resources/OnDemand_submit_Jupyter_batch_job.png)
*<p style="text-align: center;"> Figure: Launching Jupyter. </p>*
Wait until your job transitions from Queued to Running, and then connect to Jupyter.

NOTE: To run a Jupyter cell
- Type your command into the cell
- To execute the command:
    - Press **shift+return**
