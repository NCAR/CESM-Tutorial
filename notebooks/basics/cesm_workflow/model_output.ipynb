# Model Output



If the model run is successful, the CESM netCDF history files are automatically moved to a short term archive (`$DOUT_S_ROOT`) by `case.st_archive`.

![Case Submit Image](../../../images/basics/CESM2_short_term_archive.png)


For the current tutorial on Derecho, the paths are:
- `$SRCROOT`     = `/glade/u/home/$USER/code/my_cesm_code`
- `$CASEROOT`    = `/glade/u/home/$USER/cases/$CASE`
- `$EXEROOT`     =`/glade/derecho/scratch/$USER/$CASE/bld` 
- `$RUNDIR`      =`/glade/derecho/scratch/$USER/$CASE/run` 
- `$DOUT_S_ROOT` =`/glade/derecho/scratch/$USER/archive/$CASE` 

**Notes:**
- If a model run was unsuccessful, the output remains in the Run Directory (`$RUNDIR`) and the short term archive is not created.
- Both `$RUNDIR` and `$DOUT_S_ROOT` are in the GLADE scratch file space. This space is scrubbed and files deleted after a number of days (i.e., 180 days by default). Thus, it is a good idea to move your model output files from the short term archive to a more permanent location as soon as you are able.
