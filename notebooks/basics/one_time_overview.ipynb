# One-time Setup Exercise


The first lab exercise is to download the CESM from Github and explore the code. 

The lab exercise is estimated to take around 30-60 minutes, depending on your previous experience with NCAR HPC and Github.

## Learning Objectives

<div class="alert alert-info" style="text-align: left;">
    
- Download CESM2 to your workspace with Git commands.
- Explore your CESM2 workspace for CESM Code.
    
</div>

If you are unfamiliar with logging into NCAR HPC resources, please go back to [section](link) to learn more about logging into Cheyenne.

Github and Git software are used as development and management tools for supporting and distributing CESM. This tutorial is not intended to be a Github or Git software tutorial, so we will provide the basic git commands needed to download CESM. If you want to learn more about using these tools go to [section](link).
