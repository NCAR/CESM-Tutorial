# Examine History Files


Having successfully completed your first month of the new CESM B1850 case, you will now examine the history files that have been transfered to the Archive directory. You will be using the NetCDF viewer Ncview to look at the monthly average values for the single month. For this exercise you will be looking at the output from the Community Atmospheric Model (CAM) component of CESM. However, the Ncview tool can be applied to other components, such as the Community Land Model (CLM). In this exercise we will:

- Step 1. Explore the b1850.basics Archive directory.
- Step 2. Open the b1850.basics CAM monthly history file using Ncview.
- Step 3. Examine Average Monthly Surface Temperature.
- Step 4. Examine Average Monthly Precipitation.

## Step 3. Examine Average Monthly Average Surface Temperature

Using Ncview with the b1850.basics January 0001 CAM monthly average values file loaded, examine the Surface Temperature variable. The new spectral element grids in the atmosphere model are unstructured, so data will be missing a dimension in this exercise.

<div class="alert alert-info">
<strong>Select the Surface Temperature Variable: </strong><br><br>

Click (and hold the button down) on the **(225) 1d vars** button shown above. This will pull up a list of all 1d vars on the b1850.basics January 0001 CAM history file. This is a long list but it is in alphabetical order. You are looking for the `TS` variable.  
<br>

<strong>Two dimensional data is one dimensional: </strong><br>

By default, the unstructured grids in the atmosphere model now show two dimensional data as a single dimension. You will see in this plot of surface temperature that the Y axis shows temperature, but the X axis is dimensioned "ncol". This is the number of columns over the entire globe in your atmospheric simulation.
    
<br>

![NCView TS](../../../images/basics/ncview.cesm3.ts.png)
*<p style="text-align: center;"> Figure: NCView b1850.basics Average TS Jan 0001 </p>*
    
<strong>Look at Latitude and Longitude: </strong><br>

One way to understand how two dimensional data is shown as a single dimension is to look at the latitude and longitude fields in your file. Again, click (and hold the button down) on the **(225) 1d vars** button. Then choose to plot the "lat" variable or the "lon" variable. This is a list of the latitude for each "ncol" column in your history output or the longitude for each column. Again, this is the default, or "native" grid output for the unstructured grid in the atmospheric model.

![NCView TS](../../../images/basics/ncview.cesm3.latlon.png)
*<p style="text-align: center;"> Figure: NCView b1850.basics latitude and longitude Jan 0001 </p>*

NCAR scientists don't interact with the native grid data in CESM3 as much as we did for previous versions. It is more common to configure your experiment to interpolate this data to a 2 dimensional lat-lon grid for history files. But, this requires more advanced experiment modifications. Instructions can be found in the <a href="https://ncar.github.io/CESM-Tutorial/notebooks/namelist/exercises/exercise_cam_interpolate.html">Interpolate CAM Output</a> section of the tutorial.
