# Troubleshooting runtime errors

If your case fails during execution, there are several places to look for clues about what went wrong.

Some of the most useful sources of information are:

- **Log files** (`*.log.*`), which often contain the error message or stack trace that caused the failure.
- **CaseStatus**, which records the progress of the workflow and indicates which stage of the workflow failed.
- **Batch job output** (for example, `*.o*` and `*.e*` files on HPC systems), which may include scheduler messages or runtime errors.
- **Debug mode**, which provides additional diagnostic information and can make it easier to identify the source of a problem.

As you gain experience with CESM, you'll find that the log files are usually the best place to start when diagnosing runtime errors.
