1. For a branch run, we need to set several variables related to the reference case in `env_run.xml`. To find variables applied to reference case (Branch and Hybrid runs), use ./xmlquery -p REF.
    
2. Ocean CO2 need to be changed in `env_run.xml`. Atmospehre CO2 and methane are controlled by the variables `co2vmr` and `ch4vmr` respectively in the atmosphere model namelist file. You can check the default values of these variables in your `$CASEROOT/CaseDocs/atm_in`.

    Make modifications to the namelist file using `user_nl_cam`. Then run `./preview_namelist`. Check if your changes were implemented in your run directory (`$RUNDIR`) and in `$CASEROOT/CaseDocs/atm_in`. 
    Learn the [basics of namelist modification here](../../../namelist/overview.ipynb). We will dive into the details in the next chapter. 

3. Make sure to update the walltime to match the run length.
    
4. Don't forget to put restart files into the run directory `$RUNDIR`.

1. From the script directory, create your new case:
    
```
    ./create_newcase --case ~/cases/b.day2.2 --res f19_g17 --compset B1850
```
<br>
    
2. In your case directory, change the run type to branch, and change the associated RUN_* variables:

```
    cd ~/case/b.day2.2/
    ./xmlchange RUN_TYPE=branch
    ./xmlchange RUN_REFCASE=b.day2.1,RUN_REFDATE=0001-02-01,GET_REFCASE=FALSE
```
<br>    
    
3. Set run time variables: 
```
    ./xmlchange STOP_OPTION=nmonths,STOP_N=1,RESUBMIT=1,JOB_WALLCLOCK_TIME=02:00:00
```    
<br>

    
4. Change CO2 level in `env_run.xml`:
```
    ./xmlchange CCSM_CO2_PPMV=569.4
```
<br>    

5. Set up the case, put the initial and restart files in the `$RUNDIR`:
    
```
    ./case.setup
    cp /glade/scratch/$LOGNAME/archive/b.day2.1/rest/0001-02-01-00000/*  /glade/scratch/$LOGNAME/b.day2.2/run/
```
<br>  
   
6. Make changes to the atmosphere namelist. In `user_nl_cam`, add the following lines:
   
    
```
    co2vmr=569.4e-6
    ch4vmr=1583.2e-9
    inithist=‘MONTHLY‘ 
``` 
<br>

7. Run `preview_namelist` to resolve the namelist modifications. 
```
    ./preview_namelist
``` 
Check if the changes were implemented in $CASEROOT/CaseDocs/atm_in
    
<br>
8. Carry on to build and submit the run:
    
```
      qcmd -- ./case.build (qcmd is used on Cheyenne only)
      ./case.submit
```
<br>    

After the job completes, go to the short term archive space and explore.

</details>
</div>
