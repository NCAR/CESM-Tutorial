# Overview



The `SourceMods` directory contains component-specific subdirectories. Common examples include:

| Directory | Component |
|------------|------------|
| `SourceMods/src.cam` | Atmosphere (CAM) |
| `SourceMods/src.clm` | Land (CLM) |
| `SourceMods/src.cice` | Sea Ice (CICE) |
| `SourceMods/src.mom` | Ocean (MOM) |
| `SourceMods/src.mosart` | River Transport (MOSART) |

Using `SourceMods` allows you to keep your modifications isolated within a case, making it easier to track changes, reproduce experiments, and update CESM without altering the original source code.

# Overview


While many CESM customizations can be accomplished through namelist settings, some changes require modifying the model source code directly. These changes are referred to as **source modifications**.

When making source modifications, the recommended practice is **not** to edit files in the CESM source tree directly. Instead, copy the file you wish to modify into the appropriate `SourceMods` directory within your case directory and make your changes there.

For example, suppose you want to modify a CAM source file in a case called `case01`. You would copy the file into:

```text
case01/SourceMods/src.cam
```

and make your modifications in that location.

During the build process, CESM automatically uses files in the `SourceMods` directories in place of the corresponding files from the CESM source tre

The **steps**  when making a **source code modification** are:

- In the **CESM code**, find the subroutine you want to modify.
- Copy this subroutine to the relevant **SourceMods** directory within your **case directory**
- Make your **modifications**
- **Recompile** and **run** the model
