# Geoscience Data Exchange (GDEX)

Publicly released CESM data is available via the Geoscience Data Exchange (GDEX). Timeseries data in both CESM standard format and CMIP format are available. Registration is not necessary to download files, but users can login using their ORCID ID or GLOBUS for customized data downloads.
