# CESM Working Groups

A defining feature of CESM is its strong commitment to community involvement in both model development and scientific governance. CESM development is organized around a set of **Working Groups** that provide scientific leadership, coordinate model development, and help guide the future direction of the project.

Working Groups are responsible for advancing specific model components or coordinating cross-cutting scientific and technical activities across CESM. Each group develops priorities and work plans that align with the broader goals of CESM while helping ensure that the model remains a state-of-the-art Earth system modeling framework.

Each Working Group is led by one or more Co-Chairs and operates under the guidance of the CESM Scientific Steering Committee (SSC). Working Groups establish their own scientific priorities and development plans while contributing to the overall CESM mission.

**We welcome and strongly encourage anyone interested to join the relevant CESM working groups to participate more fully in the CESM project and model development.** 

CESM is a community model, and participation from users and developers is essential to its success. We strongly encourage anyone interested in CESM science, development, or applications to join the relevant Working Groups and become involved in the project.
